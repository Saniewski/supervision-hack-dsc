```python
from lxml import html
import requests

links = []
for i in range(1):
    text = requests.get(f"https://top-ogloszenia.net/11-0/page{i}.html").text
    tree = html.fromstring(text)
    print('jestem tutaj')
    print(text)
    print('teraz tutaj')
    print(tree)
    for j in range(1):
        xpath_expr = '/html/body/div[2]/div[4]/div[{}]/div/div[2]/h4/a/@href'.format(j)
        if len(tree.xpath(xpath_expr)) > 0:
            print('po drzewie')
            print(j, tree.xpath(xpath_expr))
            links.append(tree.xpath(xpath_expr)[0])
    if len(links) > 0:
        break
print(links)

with open('data/topogl_work_hrefs.csv', 'w') as f:
    for link in links:
        f.write(link + '\n')
```
