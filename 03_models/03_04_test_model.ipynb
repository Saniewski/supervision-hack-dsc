```python
import pickle
import pandas as pd

# df = pd.read_json('../data/labeled_data_for_supervised.json')
df = pd.read_json('path_to_test_file')
```

```python
phrases = []

with open('../data/suspicious_phrases.txt') as f:
    for line in f.readlines():
        phrases.append(line.strip())

df['is_suspicious'] = df['text'].apply(lambda x : 1 if any([phrase in x for phrase in phrases]) else 0)
```

```python
tfidf = pickle.load(open("TFIDF_serialized.pickle", 'rb'))
data = pd.DataFrame(tfidf.transform(df['text']).todense(), columns = tfidf.get_feature_names_out()).astype('float32')
data
```

```python
from joblib import load
clf = load('random_forest_model.joblib')
```

```python
X = data
X['is_suspicious'] = df['is_suspicious']
y = df['label']
```

```python
pred = clf.predict(X)
```

```python
from sklearn.metrics import confusion_matrix, f1_score

print(f1_score(y, pred))
confusion_matrix(y, pred)
```
